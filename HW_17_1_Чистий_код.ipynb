І «product_name», і «product_category» є рядками, що представляють численні, але обмежені унікальні назви продуктів і категорії відповідно.

Напишіть код для ефективного перетворення стовпців «product_name» і «product_category» на категорійний тип для ефективного використання пам’яті.
Спробуйте вирішити це завдання, мінімізуючи використання пам'яті та зберігаючи читабельність коду. Крім того, виміряйте використання пам’яті до та після перетворення, щоб продемонструвати підвищення ефективності.

Порівняйте кількість памʼяті, яку займає наданий набір даних і новий, створений Вами. В лекції був приклад. В скільки разів менше памʼяті займає новий датафрейм?

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})
````

````python
import pandas as pd
import numpy as np

# --- helper: memory in MB ---
def memory_mb(dataframe: pd.DataFrame) -> float:
    return dataframe.memory_usage(deep=True).sum() / (1024 ** 2)


# 1: памʼять ДО оптимізації
mem_before = memory_mb(df)

# 2: робимо копію, щоб не змінювати оригінальний df
df_opt = df.copy()

# 3: реретворюємо на категоріальні (найбільша економія памʼяті для повторюваних рядків)
cat_cols = ["product_name", "product_category"]
df_opt[cat_cols] = df_opt[cat_cols].astype("category")

# 4: памʼять ПІСЛЯ оптимізації
mem_after = memory_mb(df_opt)

# 5: у скільки разів стало менше
reduction_factor = mem_before / mem_after

print(f"Memory before: {mem_before:.2f} MB")
print(f"Memory after:  {mem_after:.2f} MB")
print(f"Reduction:     {reduction_factor:.2f}x less memory")
````

- До оптимізації стовпці product_name та product_category були типу object і займали багато памʼяті через зберігання рядків як Python-об’єктів.

- Після конвертації у category пам’ять зменшилась, тому що тепер унікальні значення зберігаються один раз + використовується масив кодів.

- Заміри memory_usage(deep=True) показали, що новий DataFrame займає в reduction_factor разів менше пам’яті.

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
input_list = [75, 80, 85, 90, 95]
````

````python
def calculate_variance(numbers):
    """
    Дисперсія для чисел у списку
    """
    if not numbers:
        raise ValueError('Список чисел не може бути порожнім.')

    mean = sum(numbers) / len(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)
````

````python
result = calculate_variance(input_list)
print(result)
````
