–Ü—Å—Ç–æ—Ä–∏—á–Ω–æ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –≤—ñ–¥ –ø–µ—Ä—à–æ–≥–æ –∑–∞—Ö–æ–¥—É –Ω–∞ —Å–∞–π—Ç –¥–æ –ø–æ–∫—É–ø–∫–∏ —Å—Ç–∞–Ω–æ–≤–∏–≤ $\mu_0 = 123.7$ —Ö–≤–∏–ª–∏–Ω.

–ö–æ–º–∞–Ω–¥–∞ –∞–Ω–∞–ª—ñ—Ç–∏–∫—ñ–≤ –≤–∏–ø–∞–¥–∫–æ–≤–æ –æ–±—Ä–∞–ª–∞ –¥–∞–Ω—ñ **7 –ø–æ–∫—É–ø—Ü—ñ–≤ –ø—ñ—Å–ª—è —Ä–µ–¥–∏–∑–∞–π–Ω—É**:
`128, 135, 121, 142, 126, 151, 123`

–ó –¥–æ–ø–æ–º–æ–≥–æ—é –Ω–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö –∑ º—è—Å—É–π—Ç–µ, —á–∏ –∑–±—ñ–ª—å—à–∏–≤—Å—è —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –ø–æ–∫—É–ø–∫–∏ –ø—ñ—Å–ª—è —Ä–µ–¥–∏–∑–∞–π–Ω—É.

–î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ–Ω–∞–π—Ç–µ –Ω–∞—Å—Ç—É–ø–Ω—ñ 6 –∫—Ä–æ–∫—ñ–≤. –ü—Ä–∞–≤–∏–ª—å–Ω–µ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –∑ –∫—Ä–æ–∫—ñ–≤ –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è –≤ 1 –±–∞–ª.

1. –ó–∞–ø–∏—à—ñ—Ç—å –Ω—É–ª—å–æ–≤—É —Ç–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É –≥—ñ–ø–æ—Ç–µ–∑–∏ —ñ –≤–∏–∑–Ω–∞—á—Ç–µ —Ç–∏–ø —Ç–µ—Å—Ç—É.
2. –û–±—á–∏—Å–ª—ñ—Ç—å –≤–∏–±—ñ—Ä–∫–æ–≤—ñ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏: $\bar x$, $s$, $n$.
3. –û–±–µ—Ä—ñ—Ç—å —Ç–∏–ø —Ç–µ—Å—Ç—É —Ç–∞ –≤–∏–∫–æ–Ω–∞–π—Ç–µ –π–æ–≥–æ (–∑–Ω–∞–π–¥—ñ—Ç—å –∫—Ä–∏—Ç–∏—á–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–µ—Å—Ç—É, —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É —Ç–µ—Å—Ç—É —Ç–∞ —Ä-–∑–Ω–∞—á–µ–Ω–Ω—è) –±—É–¥—å-—è–∫–∏–º –ø–∞—Å—É—é—á–∏–º —Å–ø–æ—Å–æ–±–æ–º, —â–æ–± –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –≥—ñ–ø–æ—Ç–µ–∑—É –Ω–∞ —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ $\alpha = 0.10$.
4. –ü—Ä–∏–π–º—ñ—Ç—å —Ä—ñ—à–µ–Ω–Ω—è, —á–∏ –≤—ñ–¥—Ö–∏–ª—è—î—Ç–µ –≤–∏ –≥—ñ–ø–æ—Ç–µ–∑—É $H_0$ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ p-value.
5. –ù–∞–ø–∏—à—ñ—Ç—å –≤–∏—Å–Ω–æ–≤–æ–∫: —á–∏ —Å–ø—Ä–∞–≤–¥—ñ —Ä–µ–¥–∏–∑–∞–π–Ω —Å–∞–π—Ç—É –ø–æ–¥–æ–≤–∂–∏–≤ —á–∞—Å –¥–æ –ø–æ–∫—É–ø–∫–∏?
6. –ß–∏ –∑–º—ñ–Ω–∏—Ç—å—Å—è –≤–∞—à–µ —Ä—ñ—à–µ–Ω–Ω—è –ø—Ä–∏ –∑–º—ñ–Ω—ñ —Ä—ñ–≤–Ω—è –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ –Ω–∞ $\alpha = 0.05$.


**–ö–†–û–ö 1: –ó–∞–ø–∏—à—ñ—Ç—å –Ω—É–ª—å–æ–≤—É —Ç–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É –≥—ñ–ø–æ—Ç–µ–∑–∏ —ñ –≤–∏–∑–Ω–∞—á—Ç–µ —Ç–∏–ø —Ç–µ—Å—Ç—É**.

- –í—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –∑–∞–∑–Ω–∞—á–∏–º–æ, —â–æ —ñ—Å—Ç–æ—Ä–∏—á–Ω–æ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –≤—ñ–¥ –ø–µ—Ä—à–æ–≥–æ –∑–∞—Ö–æ–¥—É –Ω–∞ —Å–∞–π—Ç –¥–æ –ø–æ–∫—É–ø–∫–∏ —Å—Ç–∞–Ω–æ–≤–∏–≤  ùúá0=123.7  —Ö–≤–∏–ª–∏–Ω

- –ù—É–ª—å–æ–≤–∞ –≥—ñ–ø–æ—Ç–µ–∑–∞: –∑–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º –º–∏ –ø—Ä–∏–ø—É—Å–∫–∞—î–º–æ, —â–æ –Ω—ñ—á–æ–≥–æ –Ω–µ –∑–º—ñ–Ω–∏–ª–æ—Å—å
ùêª0: ùúá = 123.7

- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞ –≥—ñ–ø–æ—Ç–µ–∑–∞: –æ—Å–∫—ñ–ª—å–∫–∏ –∫–æ–º–∞–Ω–¥–∞ –ø—ñ–¥–æ–∑—Ä—é—î, —â–æ —á–∞—Å –∑–±—ñ–ª—å—à–∏–≤—Å—è, —Ç–æ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ
Ha‚Äã: Œº>123.7

–¢–ò–ü —Ç–µ—Å—Ç—É: –≤ –Ω–∞—à–æ–º—É –≤–∏–ø–∞–¥–∫—É —Ç–µ—Å—Ç –ü–†–ê–í–û–°–¢–û–†–û–ù–ù–Ü–ô (–æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π), –æ—Å–∫—ñ–ª—å–∫–∏ –≤ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ñ–π –≥—ñ–ø–æ—Ç–∏–∑—ñ –º–∏ –ø—Ä–∏–ø—É—Å–∫–∞—î–º–æ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è —á–∞—Å—É.


**–ö–†–û–ö 2: –û–±—á–∏—Å–ª—ñ—Ç—å –≤–∏–±—ñ—Ä–∫–æ–≤—ñ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:  ùë•¬Ø ,  ùë† ,  ùëõ** .

```python
import numpy as np
```

```python
# –∑–∞–Ω–µ—Å–µ–º–æ –¥–∞–Ω—ñ –≤–∏–±—ñ—Ä–∫–∏ –≤ –º–∞—Å–∏–≤
x = np.array([128, 135, 121, 142, 126, 151, 123])
```

```python
# –æ–±—á–∏—Å–ª–∏–º–æ –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏

n = x.size                          # —Ä–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏ (–∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤ –º–∞—Å–∏–≤—ñ)
x_bar = x.mean()                    # —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –ø–æ –≤–∏–±—ñ—Ä—Ü—ñ
s = x.std(ddof=1)                   # –≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è (s), ddof=1! (–Ω–∞—Å–∫—ñ–ª—å–∫–∏ –∞–æ–Ω–∏ —Ä–æ–∑–∫–∏–¥–∞–Ω—ñ –Ω–∞–≤–∫–æ–ª–æ –Ω–∞—à–æ–≥–æ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ)

print("n =", n)
print("xÃÑ =", x_bar)
print("s =", s)
```

**–ö–†–û–ö 3 : –û–±–µ—Ä—ñ—Ç—å —Ç–∏–ø —Ç–µ—Å—Ç—É —Ç–∞ –≤–∏–∫–æ–Ω–∞–π—Ç–µ –π–æ–≥–æ (–∑–Ω–∞–π–¥—ñ—Ç—å –∫—Ä–∏—Ç–∏—á–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–µ—Å—Ç—É, —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É —Ç–µ—Å—Ç—É —Ç–∞ —Ä-–∑–Ω–∞—á–µ–Ω–Ω—è) –±—É–¥—å-—è–∫–∏–º –ø–∞—Å—É—é—á–∏–º —Å–ø–æ—Å–æ–±–æ–º, —â–æ–± –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ –≥—ñ–ø–æ—Ç–µ–∑—É –Ω–∞ —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ  ùõº=0.10** .


–ù–∞—Ä–∞–∑—ñ –º–∏ –∑–Ω–∞—î–º–æ:
- –Ω–∞—à–∞ –≤–∏–±—ñ—Ä–∫–∞ –º–∞–ª–∞: n=7

- —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –ø–æ–ø—É–ª—è—Ü—ñ—ó –Ω–µ–≤—ñ–¥–æ–º–µ

- –º–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –≤–∏–±—ñ—Ä–∫–æ–≤–µ s

–¶–µ –æ–∑–Ω–∞—á–∞—î: –Ω–∞—à –≤–∏–±—ñ—Ä –≤ –¥–∞–Ω–æ–º—É –≤–∏–ø–∞–¥–∫—É - —Ü–µ –æ–¥–Ω–æ–≤–∏–±—ñ—Ä–∫–æ–≤–∏–π t-test (Student‚Äôs t-test)

–ù–µ z-test,  –±–æ: –≤–∏–±—ñ—Ä–∫–∞ –º–µ–Ω—à–µ 30 —Ç–∞ ùúé –Ω–µ–≤—ñ–¥–æ–º–µ.

–¢–û–ë–¢–û –ú–ê–Ñ–ú–û –ü–†–ê–í–û–°–¢–û–†–û–ù–ù–Ü–ô t-—Ç–µ—Å—Ç.

–î–ª—è –Ω—å–æ–≥–æ –≤–∏—Ä–ø–∞—Ö—É—î–º–æ **degrees of freedom**: Df = n - 1 = 7 - 1 = 6

```python
# –ó–ù–ê–ô–î–ï–ú–û –ö–†–ò–¢–ò–ß–ù–ï –ó–ù–ê–ß–ï–ù–ù–Ø –ø—Ä–∏ ùõº=0.10

from scipy import stats

alpha = 0.10
df = n - 1

t_critical = stats.t.ppf(1 - alpha, df)

print("df =", df)
print("–ö—Ä–∏—Ç–∏—á–Ω–µ t-–∑–Ω–∞—á–µ–Ω–Ω—è =", t_critical)
```

```python
# –ó–ù–ê–•–û–î–ò–ú–û t-–°–¢–ê–¢–ò–°–¢–ò–ö–£ (–Ω–∞ —Å–∫—ñ–ª—å–∫–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏—Ö –ø–æ–º–∏–ª–æ–∫ –Ω–∞—à–µ —Å–µ—Ä–µ–¥–Ω—î –±—ñ–ª—å—à–µ –∑–∞ —ñ—Å—Ç–æ—Ä–∏—á–Ω–µ)

mu_0 = 123.7

t_stat = (x_bar - mu_0) / (s / np.sqrt(n))

print("t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ =", t_stat)
```

```python
# –ó–ù–ê–•–û–î–ò–ú–û p-value (–≤—Ä–∞—Ö–æ–≤—É—é—á–∏ —Ç–µ—Å—Ç –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π)

p_value = 1 - stats.t.cdf(t_stat, df)

print("p-value =", p_value)
```

**–ö–†–û–ö 5: –ù–∞–ø–∏—à—ñ—Ç—å –≤–∏—Å–Ω–æ–≤–æ–∫: —á–∏ —Å–ø—Ä–∞–≤–¥—ñ —Ä–µ–¥–∏–∑–∞–π–Ω —Å–∞–π—Ç—É –ø–æ–¥–æ–≤–∂–∏–≤ —á–∞—Å –¥–æ –ø–æ–∫—É–ø–∫–∏**?

–û—Å–∫—ñ–ª—å–∫–∏ p-value = 0.042 —î –º–µ–Ω—à–∏–º –∑–∞ —Ä—ñ–≤–µ–Ω—å –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ Œ± = 0.10, –º–∏ –≤—ñ–¥—Ö–∏–ª—è—î–º–æ –Ω—É–ª—å–æ–≤—É –≥—ñ–ø–æ—Ç–µ–∑—É. –ù–∞ —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ 10% –æ—Ç—Ä–∏–º–∞–Ω—ñ –¥–∞–Ω—ñ –Ω–∞–¥–∞—é—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â—ñ –¥–æ–∫–∞–∑–∏ —Ç–æ–≥–æ, —â–æ –ø—ñ—Å–ª—è —Ä–µ–¥–∏–∑–∞–π–Ω—É —Å–∞–π—Ç—É —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –¥–æ –ø–æ–∫—É–ø–∫–∏ –∑–±—ñ–ª—å—à–∏–≤—Å—è.–¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ —Ä–µ–¥–∏–∑–∞–π–Ω, –π–º–æ–≤—ñ—Ä–Ω–æ, –ø—Ä–∏–∑–≤—ñ–≤ –¥–æ —Ç–æ–≥–æ, —â–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –±—ñ–ª—å—à–µ —á–∞—Å—É –º—ñ–∂ –ø–µ—Ä—à–∏–º –∑–∞—Ö–æ–¥–æ–º –Ω–∞ —Å–∞–π—Ç —ñ –∑–¥—ñ–π—Å–Ω–µ–Ω–Ω—è–º –ø–æ–∫—É–ø–∫–∏.

**–ó–∞–≤–¥–∞–Ω–Ω—è 2.**

–î–æ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–æ—ó —Ä–µ–∫–ª–∞–º–Ω–æ—ó –∫–∞–º–ø–∞–Ω—ñ—ó **23%** –¥–æ—Ä–æ—Å–ª–∏—Ö —É–ø—ñ–∑–Ω–∞–≤–∞–ª–∏ –ª–æ–≥–æ—Ç–∏–ø –∫–æ–º–ø–∞–Ω—ñ—ó. –ü—ñ—Å–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è –∫–∞–º–ø–∞–Ω—ñ—ó –≤—ñ–¥–¥—ñ–ª –º–∞—Ä–∫–µ—Ç–∏–Ω–≥—É –ø—Ä–æ–≤—ñ–≤ –æ–ø–∏—Ç—É–≤–∞–Ω–Ω—è: –∑ **1200** –≤–∏–ø–∞–¥–∫–æ–≤–æ –≤—ñ–¥—ñ–±—Ä–∞–Ω–∏—Ö –¥–æ—Ä–æ—Å–ª–∏—Ö **311** —É–ø—ñ–∑–Ω–∞–ª–∏ –ª–æ–≥–æ—Ç–∏–ø.

–ü–µ—Ä–µ–≤—ñ—Ä—Ç–µ –Ω–∞ —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ **3%** ($\alpha=0.03$), —á–∏ –¥–∞—é—Ç—å —Ü—ñ –¥–∞–Ω—ñ –¥–æ—Å—Ç–∞—Ç–Ω—ñ –ø—ñ–¥—Å—Ç–∞–≤–∏ —Å—Ç–≤–µ—Ä–¥–∂—É–≤–∞—Ç–∏, —â–æ **—Ç–µ–ø–µ—Ä –±—ñ–ª—å—à–µ –Ω—ñ–∂ 23%** –¥–æ—Ä–æ—Å–ª–∏—Ö —É–ø—ñ–∑–Ω–∞—é—Ç—å –ª–æ–≥–æ—Ç–∏–ø –∫–æ–º–ø–∞–Ω—ñ—ó. –î–ª—è —Ä–æ–∑–≤ º—è–∑–∫—É –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `statsmodels`.

–ó—Ä–æ–±—ñ—Ç—å –≤–∏—Å–Ω–æ–≤–æ–∫, —á–∏ –∑—Ä–æ—Å–ª–∞ –≤–ø—ñ–∑–Ω–∞–≤–∞–Ω—ñ—Å—Ç—å –ª–æ–≥–æ—Ç–∏–ø—É.

–î–æ–¥–∞—Ç–∫–æ–≤–æ, –æ–±—á–∏—Å–ª—ñ—Ç—å –¥–æ–≤—ñ—Ä—á–∏–π —ñ–Ω—Ç—Ä–µ–≤–∞–ª –Ω–∞ –∑–∞–¥–∞–Ω–æ–º—É —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ —ñ –ø—Ä–æ—ñ–Ω—Ç–µ—Ä–ø—Ä–µ—Ç—É–π—Ç–µ —Ç–µ–∫—Å—Ç–æ–º - —è–∫ –≤—ñ–Ω –¥–æ–¥–∞—Ç–∫–æ–≤–æ –ø–æ—è—Å–Ω—é—î –ø—Ä–∏–π–Ω—è—Ç–µ –Ω–∞–º–∏ —Ä—ñ—à–µ–Ω–Ω—è?

**–ö–†–û–ö 1: –°—Ñ–æ—Ä–º—É—î–º–æ –Ω—É–ª—å–æ–≤—É —Ç–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É –≥—ñ–ø–æ—Ç–µ–∑–∏**.

–í –Ω–∞—Å —ñ—Å—Ç–æ—Ä–∏—á–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è p (–Ω—É–ª—å–æ–≤–µ) = 0.23 (—Ä–∞–Ω—ñ—à–µ 23% –≤–ø—ñ–∑–Ω–∞–≤–∞–ª–∏ –ª–æ–≥–æ –∫–æ–º–ø–∞–Ω—ñ—ó).

–ù—É–ª—å–æ–≤–∞ –≥—ñ–ø–æ—Ç–µ–∑–∞: H0‚Äã: p=0.23 (–ø—Ä–∏–ø—É—Å–∫–∞—î–º–æ, —â–æ –∫–∞–º–ø–∞–Ω—ñ—è –Ω—ñ—á–æ–≥–æ –Ω–µ –∑–º—ñ–Ω–∏–ª–∞)
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞ –≥—ñ–ø–æ—Ç–µ–∑–∞: Ha‚Äã: p>0.23 (–ø—ñ—Å–ª—è –∫–∞–º–ø–∞–Ω—ñ—ó –ª—é–¥–∏ —Å—Ç–∞–ª–∏ –∫—Ä–∞—â–µ –≤–ø—ñ–∑–Ω–∞–≤–∞—Ç–∏ –ª–æ–≥–æ).

–¢–∏–ø —Ç–µ—Å—Ç—É: –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç –¥–ª—è —á–∞—Å—Ç–∫–∏.

–†—ñ–≤–µ–Ω—å –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ: Œ±=0.03.

**–ö–†–û–ö 2: –ü—ñ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–∏—Ö –¥–ª—è —Ç–µ—Å—Ç—É —á–∞—Å—Ç–∫–∏**.




```python
# –ö–†–û–ö 2

successes = 311      # –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–∞–∑ –≤–ø—ñ–∑–Ω–∞–ª–∏ –ª–æ–≥–æ
nobs = 1200          # —Ä–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏
p0 = 0.23            # —ñ—Å—Ç–æ—Ä–∏—á–Ω–∞ —á–∞—Å—Ç–∫–∞
alpha = 0.03         # —Ä—ñ–≤–µ–Ω—å –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ

# –≤–∏–±—ñ—Ä–∫–æ–≤–∞ —á–∞—Å—Ç–∫–∞ (–¥–ª—è —Ä–æ–∑—É–º—ñ–Ω–Ω—è, –Ω–µ –¥–ª—è —Ç–µ—Å—Ç—É)
p_hat = successes / nobs

print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å —É—Å–ø—ñ—Ö—ñ–≤ (x):", successes)
print("–†–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏ (n):", nobs)
print("–Ü—Å—Ç–æ—Ä–∏—á–Ω–∞ —á–∞—Å—Ç–∫–∞ (p0):", p0)
print("–í–∏–±—ñ—Ä–∫–æ–≤–∞ —á–∞—Å—Ç–∫–∞ (pÃÇ):", p_hat)
```

**–ö—Ä–æ–∫ 3: –í–∏–∫–æ–Ω–∞—î–º–æ z-—Ç–µ—Å—Ç –¥–ª—è —á–∞—Å—Ç–∫–∏ —Ç–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∏–º–æ –º–∏–Ω—É–ª—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏**

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
# –≤–∏–∫–æ–Ω—É—î–º–æ z-test –¥–ª—è –ø—Ä–æ–ø–æ—Ä—Ü—ñ—ó
z_stat, p_value = proportions_ztest(
    count=successes,     # –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–ø—ñ–∑–Ω–∞–Ω—å
    nobs=nobs,           # —Ä–æ–∑–º—ñ—Ä –≤–∏–±—ñ—Ä–∫–∏
    value=p0,            # p0 = 0.23
    alternative='larger' # Ha: p > p0 (–ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç–µ—Å—Ç)
)

print("="*60)
print("Z-TEST –î–õ–Ø –ü–†–û–ü–û–†–¶–Ü–á (statsmodels)")
print("="*60)
print(f"z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ = {z_stat:.4f}")
print(f"p-value      = {p_value:.6f}")
print(f"alpha        = {alpha}")
```

**–ö–†–û–ö 4: –ü—Ä–∏–π–º–∞—î–º–æ —Ä—ñ—à–µ–Ω–Ω—è –Ω–∞ –æ—Å–Ω–æ–≤—ñ –¥–∞–Ω–Ω–∏—Ö**

–ó –º–∏–Ω—É–ª–æ–≥–æ –∫—Ä–æ–∫—É –±–∞—á–∏–º–æ:
- z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ = 2.3058
- p-value = 0.01056
- —Ä—ñ–≤–µ–Ω—å –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ  Œ±=0.03
- —Ç–µ—Å—Ç: –ø—Ä–∞–≤–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π

–ü–∞–º—è—Ç–∞—î–º–æ
- —è–∫—â–æ p-value ‚â§ Œ± ‚Üí –≤—ñ–¥—Ö–∏–ª—è—î–º–æ H0‚Äã
- —è–∫—â–æ p-value > Œ± ‚Üí –Ω–µ –≤—ñ–¥—Ö–∏–ª—è—î–º–æ H0

–£ –Ω–∞—Å: *0.01056 < 0.03*, **—Ç–æ–±—Ç–æ –º–∏ –≤—ñ–¥—Ö–∏–ª—è—î–º–æ –Ω—É–ª—å–æ–≤—É –≥—ñ–ø–æ—Ç–µ–∑—É —ñ —Ä–æ–±–∏–º–æ –≤–∏—Å–Ω–æ–≤–æ–∫, —â–æ –Ω–∞ —Ä—ñ–≤–Ω—ñ –∑–Ω–∞—á—É—â–æ—Å—Ç—ñ 3% –æ—Ç—Ä–∏–º–∞–Ω—ñ –¥–∞–Ω—ñ –Ω–∞–¥–∞—é—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω–æ –∑–Ω–∞—á—É—â—ñ –¥–æ–∫–∞–∑–∏ —Ç–æ–≥–æ, —â–æ —á–∞—Å—Ç–∫–∞ –¥–æ—Ä–æ—Å–ª–∏—Ö, —è–∫—ñ –≤–ø—ñ–∑–Ω–∞—é—Ç—å –ª–æ–≥–æ—Ç–∏–ø, –ø–µ—Ä–µ–≤–∏—â—É—î 23%**.
